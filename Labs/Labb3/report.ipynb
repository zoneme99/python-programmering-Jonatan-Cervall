# Analys
- Nedanför finns det 3 linjer med relativt lika lutning k i grafen. Dessa 3 har klassifierat punkerna väldigt lika. Min linje (blå) har bara en punkts skillnad jämfört med grön och röd.
- gul linje har en extrem stor lutning, detta gör att den inverterar klassifiseringen. Ration Left/Right på gula linjen är lika stor som min linje däremot ser man på grafen att punkterna är på fel plats. Gula punkter(Left) är på höger sidan och tvärtom. Detta är pga att färgerna är klassifiserade efter gula linjen
- För att förhindra avvikande klassifiseringslinjer skulle jag sätta ett intervall på k. 

$-0.5 <= k <= -2$

## Resonemang
Varför jag anser det är bra med ett intervall och filtrera bort avvikande linjer är av den anledningen att jag vill inte ha en inverterad klassifisering. Alla andra linjer (blå,grön,röd) gick parallellt med de 2 data kluster.
Gula linjen dela upp kluster diagonalt. Eftersom datan ser exakt likdan ut så kan det bli problem senare med ny data som klassifieras med linje gul.

```python
import Labb3

myline = Labb3.line(-1, 0, "unlabelled_data.csv")
myline.display("blue")
line1 = Labb3.line(-0.489, 0, "unlabelled_data.csv")
line1.display("green")
line2 = Labb3.line(-2, 0.16, "unlabelled_data.csv")
line2.display("red")
line3 = Labb3.line(800, -120, "unlabelled_data.csv")
line3.display("yellow")

Labb3.plt.show()
print(f"Blue(Left/Right): {len(myline.left)}/{len(myline.right)}")
print(f"Green(Left/Right): {len(line1.left)}/{len(line1.right)}")
print(f"Red(Left/Right): {len(line2.left)}/{len(line2.right)}")
print(f"Yellow(Left/Right): {len(line3.left)}/{len(line3.right)}")
```
